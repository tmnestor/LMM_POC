```python
import sys
from pathlib import Path

# Add project root to path so we can import benchmark functions
project_root = str(Path(responses_jsonl).parent.parent.parent.parent)
sys.path.insert(0, project_root)

from benchmark_wildreceipt import (
    SCALAR_CLASSES,
    _entity_matches,
    _extract_predicted_entities,
)

# Load predictions + embedded ground truth from JSONL
raw_df = pd.read_json(responses_jsonl, lines=True)

# Build comparison rows
rows = []
for _, row in raw_df.iterrows():
    file_name = row["file"]
    gt_entities = row.get("ground_truth", {})
    if not gt_entities or not isinstance(gt_entities, dict):
        continue

    pred_entities = _extract_predicted_entities(row["parsed"])

    all_classes = sorted(set(list(pred_entities.keys()) + list(gt_entities.keys())))

    for class_name in all_classes:
        gt_vals = gt_entities.get(class_name, [])
        pred_vals = pred_entities.get(class_name, [])

        # For scalar classes, compare as single values
        if class_name in SCALAR_CLASSES:
            gt_str = " ".join(gt_vals) if gt_vals else "(absent)"
            pred_str = " ".join(pred_vals) if pred_vals else "(absent)"
            if gt_str == "(absent)" and pred_str == "(absent)":
                match = "TP (both absent)"
            elif gt_str == "(absent)":
                match = "FP (hallucinated)"
            elif pred_str == "(absent)":
                match = "FN (missed)"
            else:
                score = _entity_matches(class_name, pred_str, gt_str)
                match = f"TP ({score:.2f})" if score else "MISMATCH"
            rows.append({
                "file": file_name,
                "class": class_name,
                "gt_value": gt_str,
                "pred_value": pred_str,
                "result": match,
            })
        else:
            # List classes — position-aware, index by index
            max_len = max(len(gt_vals), len(pred_vals))
            for i in range(max_len):
                gt_str = gt_vals[i] if i < len(gt_vals) else "(absent)"
                pred_str = pred_vals[i] if i < len(pred_vals) else "(absent)"
                if gt_str == "(absent)":
                    match = "FP (extra)"
                elif pred_str == "(absent)":
                    match = "FN (missed)"
                else:
                    score = _entity_matches(class_name, pred_str, gt_str)
                    match = f"TP ({score:.2f})" if score else "MISMATCH"
                rows.append({
                    "file": file_name,
                    "class": class_name,
                    "index": i,
                    "gt_value": gt_str,
                    "pred_value": pred_str,
                    "result": match,
                })

comparison = pd.DataFrame(rows)
print(f"{len(comparison)} comparisons across {raw_df['file'].nunique()} images")
comparison.head(20)
```

```python
# Show only mismatches and errors
errors = comparison[~comparison["result"].str.startswith("TP")]
print(f"{len(errors)} errors out of {len(comparison)} comparisons ({len(errors)/len(comparison)*100:.1f}%)")
errors
```
