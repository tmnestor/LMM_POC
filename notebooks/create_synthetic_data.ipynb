# Create Simple Synthetic Annotation Dataset

Creates a clean synthetic dataset with:
- 100 records
- 2 annotators
- Business document text using relevant terms
- Clean structure ready for testing

```python
def generate_business_text(num_words=50):
    """Generate text using business terms."""
    words = []
    for _ in range(num_words):
        if np.random.random() < 0.4:  # 40% business terms
            words.append(np.random.choice(business_terms))
        elif np.random.random() < 0.3:  # 30% numbers
            words.append(str(np.random.randint(100, 9999)))
        else:  # 30% currency amounts
            amount = np.random.uniform(10, 5000)
            words.append(f"${amount:.2f}")
    return ' '.join(words)

# Test the function
sample_text = generate_business_text(20)
print("Sample generated text:")
print(sample_text)
```

```python
def create_clean_synthetic_data():
    """Create a clean synthetic dataset with 100 records matching the required format."""
    
    # Prediction types from the image
    pred_types = [
        'supplier_a_pgs', 'header_a_pg', 'payer_a_pgs', 'address_extra', 
        'other', 'total_a_pgs', 'invDate_a_pgs', 'supplierABN_a_pgs',
        'phones_extra', 'emails_extra'
    ]
    
    # Corresponding annotator types (often match pred but can differ)
    annotator_types = [
        'supplier_a_pgs', 'header_a_pg', 'payer_a_pgs', 'address_extra',
        'other', 'other_desc_a_li', 'total_a_pg', 'invDate_a_pgs', 
        'supplierABN_a_pgs', 'phones_extra', 'emails_extra'
    ]
    
    # Create data for 5 different images with multiple blocks each
    data = []
    
    # Create 5 unique images
    for img_num in range(1, 6):
        image_name = f"1-{100 + img_num}RPXKD_1254040126_3_{img_num}"
        
        # Each image has 15-25 blocks
        num_blocks = np.random.randint(15, 26)
        
        for block_num in range(num_blocks):
            block_id = float(block_num)
            
            # Generate appropriate words based on pred type
            pred_type = np.random.choice(pred_types)
            annotator_type = np.random.choice(annotator_types)
            
            # Generate context-appropriate words
            if 'supplier' in pred_type:
                words = generate_supplier_text()
            elif 'address' in pred_type:
                words = generate_address_text()
            elif 'total' in pred_type:
                words = generate_amount_text()
            elif 'date' in pred_type or 'invDate' in pred_type:
                words = generate_date_text()
            elif 'phone' in pred_type:
                words = generate_phone_text()
            elif 'email' in pred_type:
                words = generate_email_text()
            elif 'ABN' in pred_type:
                words = generate_abn_text()
            elif 'header' in pred_type:
                words = generate_header_text()
            elif 'payer' in pred_type:
                words = generate_payer_text()
            else:  # other
                words = generate_business_text(np.random.randint(5, 20))
            
            record = {
                'image_name': image_name,
                'block_ids': block_id,
                'words': words,
                'pred': pred_type,
                'annotator': annotator_type
            }
            
            data.append(record)
    
    # Create DataFrame
    df = pd.DataFrame(data)
    
    # Sort by image_name then block_ids for consistency
    df = df.sort_values(['image_name', 'block_ids']).reset_index(drop=True)
    
    return df

def generate_supplier_text():
    """Generate supplier-related text."""
    suppliers = ['mater foundation', 'mater foundation limited', 'foundation ltd', 
                'charity foundation', 'mater hospital', 'health foundation']
    return np.random.choice(suppliers)

def generate_address_text():
    """Generate address-related text."""
    addresses = ['4a augusta street', 'east corrimal nsw 2518', 'australia', 
                '14 stratton street 14', 'newstead qld 4006', '123 main street']
    return np.random.choice(addresses)

def generate_amount_text():
    """Generate amount/total text."""
    amounts = ['amount: $30.00', '$50.00', 'total: $125.50', '$2.00', 
              f'${np.random.uniform(10, 500):.2f}']
    return np.random.choice(amounts)

def generate_date_text():
    """Generate date text."""
    dates = ['date: 18/12/2022 18/12/2022', '18/12/2022', '12/01/2023', 
            '25/06/2022', 'date: 15/03/2023']
    return np.random.choice(dates)

def generate_phone_text():
    """Generate phone text."""
    phones = ['telephone: 1800 500 422', '02 9876 5432', 'phone: 1300 123 456',
             '07 3234 5678']
    return np.random.choice(phones)

def generate_email_text():
    """Generate email text."""
    emails = ['email: foundationmater@hospital.com', 'info@foundation.org',
             'contact@charity.com.au', 'admin@mater.org.au']
    return np.random.choice(emails)

def generate_abn_text():
    """Generate ABN text."""
    abns = ['abn 9672 3184 640', 'ABN: 1234 5678 901', 'abn 9876 5432 109']
    return np.random.choice(abns)

def generate_header_text():
    """Generate header text."""
    headers = ['official donation receipt', 'tax invoice', 'receipt', 
              'donation receipt', 'payment confirmation']
    return np.random.choice(headers)

def generate_payer_text():
    """Generate payer text."""
    payers = ['carmen ter rahe', 'john smith', 'mary johnson', 'david wilson']
    return np.random.choice(payers)

print("✅ Updated function defined with realistic document structure")
```

```python
# Create the dataset
print("🚀 Creating synthetic dataset...")
df = create_clean_synthetic_data()

print(f"\n✅ Created dataset with {len(df)} records")
print(f"   Columns: {', '.join(df.columns)}")
print(f"   Shape: {df.shape}")
```

```python
# ============================================================================
# PIVOT TRANSFORMATION - MODERN PANDAS APPROACH (FIXED FOR COMMA-SEPARATED VALUES)
# ============================================================================

print("🔄 Transforming data using modern pandas pivot operations...")
print("=" * 70)

# Step 1: Show current data structure
print(f"📊 Original data structure:")
print(f"   Shape: {df.shape}")
print(f"   Columns: {list(df.columns)}")

# Step 2: Handle comma-separated annotator values (CRITICAL FIX)
print(f"\n🔧 Preprocessing: Splitting comma-separated annotator values...")

# Check if we have comma-separated values
comma_separated_count = df['annotator'].str.contains(',').sum()
print(f"   Rows with comma-separated annotators: {comma_separated_count}")

if comma_separated_count > 0:
    print("   ⚠️ Found comma-separated values - splitting them first")
    
    # Split comma-separated annotator values into separate rows
    expanded_data = []
    for _, row in df.iterrows():
        annotator_values = [ann.strip() for ann in row['annotator'].split(',')]
        for annotator in annotator_values:
            expanded_data.append({
                'image_name': row['image_name'],
                'block_ids': row['block_ids'],
                'words': row['words'],
                'pred': row['pred'],
                'annotator': annotator
            })
    
    # Create expanded dataframe
    df_expanded = pd.DataFrame(expanded_data)
    print(f"   📈 Expanded from {len(df)} to {len(df_expanded)} rows after splitting")
    
else:
    print("   ✅ No comma-separated values found - using original data")
    df_expanded = df.copy()

# Step 3: Get unique ANNOTATOR values after splitting
unique_annotator_values = df_expanded['annotator'].unique()
print(f"\n📋 Unique ANNOTATOR values (after splitting): {len(unique_annotator_values)}")
for i, annotator_val in enumerate(sorted(unique_annotator_values), 1):
    print(f"   {i:2d}. {annotator_val}")

# Step 4: Create pivot transformation using cleaned data
print(f"\n🎯 Creating pivot transformation...")
print(f"   - Index: image_name") 
print(f"   - Columns: unique ANNOTATOR values ({len(unique_annotator_values)} columns)")
print(f"   - Data placement: Direct from split ANNOTATOR values")
print(f"   - Values: words content aggregated for multiple entries")

try:
    # Pivot using the expanded dataframe with split annotator values
    transformed_df = (df_expanded.groupby(['image_name', 'annotator'])['words']
                     .apply(lambda x: ' | '.join(x) if len(x) > 1 else x.iloc[0])
                     .reset_index()
                     .pivot(index='image_name', columns='annotator', values='words')
                     .fillna('NOT_FOUND')
                     .reset_index())
    
    # Ensure ALL unique annotator values are represented as columns
    # Add any missing annotator columns with NOT_FOUND values
    for annotator in sorted(unique_annotator_values):
        if annotator not in transformed_df.columns:
            transformed_df[annotator] = 'NOT_FOUND'
    
    # Reorder columns: image_name first, then all annotators alphabetically
    column_order = ['image_name'] + sorted([col for col in transformed_df.columns if col != 'image_name'])
    transformed_df = transformed_df[column_order]
    
    print(f"\n✅ Transformation completed successfully!")
    print(f"   📊 Result shape: {transformed_df.shape}")
    print(f"   📝 Columns: {len(transformed_df.columns)} total (1 image_name + {len(transformed_df.columns)-1} annotator columns)")
    print(f"   🖼️ Images: {len(transformed_df)} unique images")
    
    # Display column information
    print(f"\n📋 New column structure:")
    print(f"   1. image_name (index column)")
    for i, col in enumerate(sorted(transformed_df.columns[1:]), 2):
        print(f"   {i:2d}. {col}")
    
    # Show sample of transformed data
    print(f"\n📄 Sample transformed data (first 3 rows, first 5 columns):")
    print("=" * 70)
    sample_cols = ['image_name'] + sorted(list(transformed_df.columns[1:]))[:4]  # First 4 annotator columns
    sample_df = transformed_df[sample_cols].head(3)
    
    for idx, row in sample_df.iterrows():
        print(f"\nImage: {row['image_name']}")
        for col in sample_cols[1:]:
            value = row[col]
            if pd.notna(value) and value != 'NOT_FOUND':
                display_value = value[:50] + "..." if len(str(value)) > 50 else value
                print(f"   {col:<25}: {display_value}")
            else:
                print(f"   {col:<25}: NOT_FOUND")
    
    # Statistics about the transformation  
    print(f"\n📊 Transformation Statistics:")
    print("=" * 70)
    total_cells = (len(transformed_df) * (len(transformed_df.columns) - 1))
    filled_cells = ((transformed_df.iloc[:, 1:] != 'NOT_FOUND').sum().sum())
    fill_rate = (filled_cells / total_cells) * 100 if total_cells > 0 else 0
    
    print(f"Total data cells: {total_cells:,}")
    print(f"Filled cells: {filled_cells:,}")
    print(f"Fill rate: {fill_rate:.1f}%")
    print(f"Empty cells: {total_cells - filled_cells:,}")
    
    # Column-wise fill statistics
    print(f"\n📈 Column fill rates (sorted by fill rate):")
    fill_stats = []
    for col in transformed_df.columns[1:]:
        col_filled = (transformed_df[col] != 'NOT_FOUND').sum()
        col_rate = (col_filled / len(transformed_df)) * 100
        fill_stats.append((col, col_filled, len(transformed_df), col_rate))
    
    # Sort by fill rate descending
    fill_stats.sort(key=lambda x: x[3], reverse=True)
    
    for col, filled, total, rate in fill_stats:
        print(f"   {col:<25}: {filled}/{total} ({rate:.1f}%)")
    
    # Show splitting examples if any occurred
    if comma_separated_count > 0:
        print(f"\n🔗 Comma-splitting examples:")
        print("=" * 70)
        split_examples = df[df['annotator'].str.contains(',')].head(3)
        for _, example in split_examples.iterrows():
            original = example['annotator']
            split_vals = [ann.strip() for ann in original.split(',')]
            print(f"   Original: '{original}'")
            print(f"   Split to: {split_vals}")
            print(f"   Words: '{example['words'][:30]}...'")
            print()
    
    print(f"\n🎉 Modern pandas pivot transformation completed successfully!")
    print(f"   ✅ Column headers: Individual ANNOTATOR values (comma-separated split)")
    print(f"   ✅ Total annotator columns: {len(unique_annotator_values)}")  
    print(f"   ✅ Aggregated multiple words using ' | ' separator")
    print(f"   ✅ Used 'NOT_FOUND' for missing combinations")
    print(f"   ✅ Maintained data integrity with {len(transformed_df)} images")

except Exception as e:
    print(f"❌ Transformation failed: {e}")
    print(f"💡 Error details: {type(e).__name__}: {str(e)}")
    transformed_df = None
    
    # Show debugging information
    print(f"\n🔍 Debugging information:")
    print(f"   DataFrame shape: {df.shape}")
    print(f"   Image name count: {df['image_name'].nunique()}")
    print(f"   Annotator value count: {df['annotator'].nunique()}")
    print(f"   Comma-separated count: {comma_separated_count}")
    print(f"   Sample annotator values:")
    for ann in df['annotator'].unique()[:5]:
        print(f"      '{ann}'")
```
