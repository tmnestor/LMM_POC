# Create Simple Synthetic Annotation Dataset

Creates a clean synthetic dataset with:
- 100 records
- 2 annotators
- Business document text using relevant terms
- Clean structure ready for testing

```python
def generate_business_text(num_words=50):
    """Generate text using business terms."""
    words = []
    for _ in range(num_words):
        if np.random.random() < 0.4:  # 40% business terms
            words.append(np.random.choice(business_terms))
        elif np.random.random() < 0.3:  # 30% numbers
            words.append(str(np.random.randint(100, 9999)))
        else:  # 30% currency amounts
            amount = np.random.uniform(10, 5000)
            words.append(f"${amount:.2f}")
    return ' '.join(words)

# Test the function
sample_text = generate_business_text(20)
print("Sample generated text:")
print(sample_text)
```

```python
def create_clean_synthetic_data():
    """Create a clean synthetic dataset with 100 records matching the required format."""
    
    # Prediction types from the image
    pred_types = [
        'supplier_a_pgs', 'header_a_pg', 'payer_a_pgs', 'address_extra', 
        'other', 'total_a_pgs', 'invDate_a_pgs', 'supplierABN_a_pgs',
        'phones_extra', 'emails_extra'
    ]
    
    # Corresponding annotator types (often match pred but can differ)
    annotator_types = [
        'supplier_a_pgs', 'header_a_pg', 'payer_a_pgs', 'address_extra',
        'other', 'other_desc_a_li', 'total_a_pg', 'invDate_a_pgs', 
        'supplierABN_a_pgs', 'phones_extra', 'emails_extra'
    ]
    
    # Create data for 5 different images with multiple blocks each
    data = []
    
    # Create 5 unique images
    for img_num in range(1, 6):
        image_name = f"1-{100 + img_num}RPXKD_1254040126_3_{img_num}"
        
        # Each image has 15-25 blocks
        num_blocks = np.random.randint(15, 26)
        
        for block_num in range(num_blocks):
            block_id = float(block_num)
            
            # Generate appropriate words based on pred type
            pred_type = np.random.choice(pred_types)
            annotator_type = np.random.choice(annotator_types)
            
            # Generate context-appropriate words
            if 'supplier' in pred_type:
                words = generate_supplier_text()
            elif 'address' in pred_type:
                words = generate_address_text()
            elif 'total' in pred_type:
                words = generate_amount_text()
            elif 'date' in pred_type or 'invDate' in pred_type:
                words = generate_date_text()
            elif 'phone' in pred_type:
                words = generate_phone_text()
            elif 'email' in pred_type:
                words = generate_email_text()
            elif 'ABN' in pred_type:
                words = generate_abn_text()
            elif 'header' in pred_type:
                words = generate_header_text()
            elif 'payer' in pred_type:
                words = generate_payer_text()
            else:  # other
                words = generate_business_text(np.random.randint(5, 20))
            
            record = {
                'image_name': image_name,
                'block_ids': block_id,
                'words': words,
                'pred': pred_type,
                'annotator': annotator_type
            }
            
            data.append(record)
    
    # Create DataFrame
    df = pd.DataFrame(data)
    
    # Sort by image_name then block_ids for consistency
    df = df.sort_values(['image_name', 'block_ids']).reset_index(drop=True)
    
    return df

def generate_supplier_text():
    """Generate supplier-related text."""
    suppliers = ['mater foundation', 'mater foundation limited', 'foundation ltd', 
                'charity foundation', 'mater hospital', 'health foundation']
    return np.random.choice(suppliers)

def generate_address_text():
    """Generate address-related text."""
    addresses = ['4a augusta street', 'east corrimal nsw 2518', 'australia', 
                '14 stratton street 14', 'newstead qld 4006', '123 main street']
    return np.random.choice(addresses)

def generate_amount_text():
    """Generate amount/total text."""
    amounts = ['amount: $30.00', '$50.00', 'total: $125.50', '$2.00', 
              f'${np.random.uniform(10, 500):.2f}']
    return np.random.choice(amounts)

def generate_date_text():
    """Generate date text."""
    dates = ['date: 18/12/2022 18/12/2022', '18/12/2022', '12/01/2023', 
            '25/06/2022', 'date: 15/03/2023']
    return np.random.choice(dates)

def generate_phone_text():
    """Generate phone text."""
    phones = ['telephone: 1800 500 422', '02 9876 5432', 'phone: 1300 123 456',
             '07 3234 5678']
    return np.random.choice(phones)

def generate_email_text():
    """Generate email text."""
    emails = ['email: foundationmater@hospital.com', 'info@foundation.org',
             'contact@charity.com.au', 'admin@mater.org.au']
    return np.random.choice(emails)

def generate_abn_text():
    """Generate ABN text."""
    abns = ['abn 9672 3184 640', 'ABN: 1234 5678 901', 'abn 9876 5432 109']
    return np.random.choice(abns)

def generate_header_text():
    """Generate header text."""
    headers = ['official donation receipt', 'tax invoice', 'receipt', 
              'donation receipt', 'payment confirmation']
    return np.random.choice(headers)

def generate_payer_text():
    """Generate payer text."""
    payers = ['carmen ter rahe', 'john smith', 'mary johnson', 'david wilson']
    return np.random.choice(payers)

print("✅ Updated function defined with realistic document structure")
```

```python
# ============================================================================
# PIVOT TRANSFORMATION - FINAL VERSION WITH COMMA-SPLITTING + EXCEPTION HANDLING
# ============================================================================

print("🔄 Final pivot transformation with comma-splitting AND exception handling...")
print("=" * 70)

# Step 1: Handle comma-separated annotator values FIRST
print(f"🔧 Step 1: Splitting comma-separated annotator values...")

comma_separated_count = df['annotator'].str.contains(',', na=False).sum()
print(f"   Rows with comma-separated annotators: {comma_separated_count}")

if comma_separated_count > 0:
    print("   ⚠️ Found comma-separated values - splitting them first")
    
    # Split comma-separated annotator values into separate rows
    expanded_data = []
    for _, row in df.iterrows():
        annotator_str = str(row['annotator']) if pd.notna(row['annotator']) else 'unknown'
        words_str = str(row['words']) if pd.notna(row['words']) else 'NOT_FOUND'
        
        # Split on commas and remove duplicates
        annotator_values = list(set([ann.strip() for ann in annotator_str.split(',') if ann.strip()]))
        
        if not annotator_values:
            annotator_values = ['unknown']
            
        for annotator in annotator_values:
            expanded_data.append({
                'image_name': str(row['image_name']) if pd.notna(row['image_name']) else 'unknown_image',
                'block_ids': row['block_ids'] if pd.notna(row['block_ids']) else 0.0,
                'words': words_str,
                'pred': str(row['pred']) if pd.notna(row['pred']) else 'unknown',
                'annotator': annotator
            })
    
    df_expanded = pd.DataFrame(expanded_data)
    print(f"   📈 Expanded from {len(df)} to {len(df_expanded)} rows after splitting")
    
else:
    print("   ✅ No comma-separated values found - using original data")
    df_expanded = df.copy()
    
    # Ensure string consistency
    df_expanded['image_name'] = df_expanded['image_name'].astype(str)
    df_expanded['words'] = df_expanded['words'].fillna('NOT_FOUND').astype(str)
    df_expanded['pred'] = df_expanded['pred'].fillna('unknown').astype(str)
    df_expanded['annotator'] = df_expanded['annotator'].fillna('unknown').astype(str)

# Step 2: Analysis of pred vs annotator mismatches (AFTER splitting)
unique_annotator_values = sorted(df_expanded['annotator'].unique())  
unique_pred_values = sorted(df_expanded['pred'].unique())

print(f"\n📋 ANNOTATOR values (column headers) AFTER splitting: {len(unique_annotator_values)}")
for i, val in enumerate(unique_annotator_values, 1):
    print(f"   {i:2d}. {val}")

print(f"\n🎯 PRED values (data placement): {len(unique_pred_values)}")
for i, val in enumerate(unique_pred_values, 1):
    print(f"   {i:2d}. {val}")

# Find mismatches between pred and annotator values
pred_set = set(unique_pred_values)
annotator_set = set(unique_annotator_values)

matched_preds = pred_set.intersection(annotator_set)
unmatched_preds = pred_set - annotator_set
orphaned_annotators = annotator_set - pred_set

print(f"\n🔍 PRED-ANNOTATOR Analysis (after comma-splitting):")
print(f"   ✅ Matched (pred ∩ annotator): {len(matched_preds)}")
for pred in sorted(matched_preds):
    print(f"      - {pred}")

print(f"\n   ⚠️ Unmatched preds (no annotator column): {len(unmatched_preds)}")
for pred in sorted(unmatched_preds):
    print(f"      - {pred}")

print(f"\n   📋 Orphaned annotators (no pred data): {len(orphaned_annotators)}")
for ann in sorted(orphaned_annotators):
    print(f"      - {ann}")

# Step 3: Create pivot with exception handling
print(f"\n🎯 Creating pivot with EXCEPTION HANDLING...")

try:
    # Initialize result dataframe with all images
    unique_images = sorted(df_expanded['image_name'].unique())
    transformed_df = pd.DataFrame({'image_name': unique_images})
    
    # Initialize ALL annotator columns (complete taxonomy) - AFTER splitting
    for col in sorted(unique_annotator_values):
        transformed_df[col] = 'NOT_FOUND'
    
    print(f"   📊 Created base dataframe: {transformed_df.shape}")
    print(f"   📝 Columns: image_name + {len(unique_annotator_values)} annotator columns")
    
    # Safe join function
    def safe_join(x):
        str_values = [str(val) for val in x if pd.notna(val)]
        return ' | '.join(str_values) if str_values else 'NOT_FOUND'
    
    # Group by image_name and pred for data placement (using expanded data)
    pred_to_words = (df_expanded.groupby(['image_name', 'pred'])['words']
                    .apply(safe_join)
                    .reset_index())
    
    print(f"   🔄 Processing {len(pred_to_words)} image+pred combinations...")
    
    # Place words data with exception handling
    placements_made = 0
    exception_count = 0
    exceptions_logged = []
    
    for _, row in pred_to_words.iterrows():
        image_name = row['image_name']
        pred_value = row['pred']
        words_content = row['words']
        
        if pred_value in unique_annotator_values:
            # NORMAL CASE: pred matches annotator column
            mask = transformed_df['image_name'] == image_name
            if mask.any():
                idx = transformed_df[mask].index[0]
                current_content = transformed_df.loc[idx, pred_value]
                if current_content == 'NOT_FOUND':
                    transformed_df.loc[idx, pred_value] = words_content
                    placements_made += 1
                else:
                    # Aggregate multiple entries
                    transformed_df.loc[idx, pred_value] = f"{current_content} | {words_content}"
                    placements_made += 1
        else:
            # EXCEPTION CASE: pred doesn't match any annotator column
            exception_count += 1
            exceptions_logged.append({
                'image': image_name,
                'pred': pred_value,
                'words_preview': words_content[:30] + "..." if len(words_content) > 30 else words_content
            })
            
            print(f"      ⚠️ EXCEPTION: pred='{pred_value}' has no matching annotator column")
    
    # Reorder columns alphabetically
    column_order = ['image_name'] + sorted([col for col in transformed_df.columns if col != 'image_name'])
    transformed_df = transformed_df[column_order]
    
    print(f"\n✅ Pivot transformation completed with comma-splitting + exception handling!")
    print(f"   📊 Result shape: {transformed_df.shape}")
    print(f"   📝 Columns: {len(transformed_df.columns)} total")
    print(f"   ✅ Successful placements: {placements_made}")
    print(f"   ⚠️ Exceptions handled: {exception_count}")
    
    if exceptions_logged:
        print(f"\n🚨 Exception Summary:")
        print("=" * 70)
        for exc in exceptions_logged[:3]:  # Show first 3 exceptions
            print(f"   Image: {exc['image']}")
            print(f"   Pred: '{exc['pred']}' → NO MATCHING COLUMN")
            print(f"   Words: '{exc['words_preview']}'")
            print()
        
        if len(exceptions_logged) > 3:
            print(f"   ... and {len(exceptions_logged) - 3} more exceptions")
    
    # Final statistics
    print(f"\n📊 Final Statistics:")
    print("=" * 70)
    total_cells = len(transformed_df) * (len(transformed_df.columns) - 1)
    filled_cells = ((transformed_df.iloc[:, 1:] != 'NOT_FOUND').sum().sum())
    fill_rate = (filled_cells / total_cells) * 100 if total_cells > 0 else 0
    
    print(f"Total data cells: {total_cells:,}")
    print(f"Filled cells: {filled_cells:,}")
    print(f"Fill rate: {fill_rate:.1f}%")
    print(f"Exception rate: {(exception_count / len(pred_to_words)) * 100:.1f}%")
    
    print(f"\n🎯 FINAL LOGIC CONFIRMED:")
    print("=" * 70)
    print(f"   ✅ Column headers: ANNOTATOR values (complete taxonomy) - COMMA-SPLIT")
    print(f"   ✅ Data placement: PRED values (with exception handling)")
    print(f"   ✅ Comma-splitting: Handled {comma_separated_count} rows")
    print(f"   ✅ Final column count: {len(unique_annotator_values)} (not 615!)")
    print(f"   ✅ Result: Shows model prediction distribution across ground truth structure")

except Exception as e:
    print(f"❌ Transformation failed: {e}")
    print(f"💡 Error details: {type(e).__name__}: {str(e)}")
    transformed_df = None
```

```python
# ============================================================================
# TEST COMMA-SEPARATED ANNOTATOR VALUES WITH CORRECTED LOGIC
# ============================================================================

print("🧪 Testing comma-separated annotator functionality WITH CORRECTED LOGIC...")
print("=" * 70)

# Create test data with comma-separated annotator values (like production)
test_data = [
    {
        'image_name': 'test_image_1',
        'block_ids': 1.0,
        'words': 'supplier foundation',
        'pred': 'supplier_a_pgs',  # This pred will be placed in supplier_a_pgs column
        'annotator': 'supplier_a_pgs,header_a_pg'  # These create the columns
    },
    {
        'image_name': 'test_image_1',
        'block_ids': 2.0,
        'words': 'total amount $100.00',
        'pred': 'total_a_pg',  # This pred will be placed in total_a_pg column
        'annotator': 'total_a_pg,total_a_pg'  # Duplicate comma-separated (creates total_a_pg column)
    },
    {
        'image_name': 'test_image_1',
        'block_ids': 3.0,
        'words': '123 main street',
        'pred': 'address_extra',  # This pred will be placed in address_extra column
        'annotator': 'address_extra'  # Single value (creates address_extra column)
    },
    {
        'image_name': 'test_image_2',
        'block_ids': 1.0,
        'words': 'invoice date 2024-01-01',
        'pred': 'invDate_a_pgs',  # This pred will be placed in invDate_a_pgs column
        'annotator': 'invDate_a_pgs,invDate_a_pgs'  # Duplicate comma-separated
    },
    {
        'image_name': 'test_image_2',
        'block_ids': 2.0,
        'words': 'ABN 12345678901',
        'pred': 'supplierABN_a_pgs',  # This pred will be placed in supplierABN_a_pgs column
        'annotator': 'supplierABN_a_pgs,address_extra'  # Mixed comma-separated
    },
    {
        'image_name': 'test_image_2',
        'block_ids': 3.0,
        'words': 'mismatched prediction',
        'pred': 'nonexistent_pred',  # This pred WON'T match any annotator column
        'annotator': 'other_extra'  # Creates other_extra column, but pred won't place data there
    }
]

# Create test dataframe
test_df = pd.DataFrame(test_data)

print(f"📋 Test data created:")
print(f"   Shape: {test_df.shape}")
print(f"   Original annotator values (for column creation):")
for i, ann in enumerate(test_df['annotator'].unique(), 1):
    print(f"      {i}. '{ann}'")
print(f"   Original pred values (for data placement):")
for i, pred in enumerate(test_df['pred'].unique(), 1):
    print(f"      {i}. '{pred}'")

print(f"\n🔧 Applying comma-separation logic...")

# Apply the same splitting logic for annotators
comma_separated_count = test_df['annotator'].str.contains(',').sum()
print(f"   Rows with commas: {comma_separated_count}")

# Split comma-separated values
expanded_test_data = []
for _, row in test_df.iterrows():
    annotator_values = list(set([ann.strip() for ann in row['annotator'].split(',')]))  # Remove duplicates
    for annotator in annotator_values:
        expanded_test_data.append({
            'image_name': row['image_name'],
            'block_ids': row['block_ids'],
            'words': row['words'],
            'pred': row['pred'],
            'annotator': annotator
        })

expanded_test_df = pd.DataFrame(expanded_test_data)

print(f"\n📈 Results after splitting:")
print(f"   Original rows: {len(test_df)}")
print(f"   Expanded rows: {len(expanded_test_df)}")
print(f"   Unique annotators after splitting (for columns): {expanded_test_df['annotator'].nunique()}")
print(f"   Individual annotator values:")
for i, ann in enumerate(sorted(expanded_test_df['annotator'].unique()), 1):
    print(f"      {i}. '{ann}'")

print(f"   Unique pred values (for data placement): {expanded_test_df['pred'].nunique()}")
for i, pred in enumerate(sorted(expanded_test_df['pred'].unique()), 1):
    print(f"      {i}. '{pred}'")

# CORRECTED LOGIC: Use annotator for columns, pred for data placement
print(f"\n🎯 Applying CORRECTED pivot logic...")
print(f"   ✅ Column headers: From ANNOTATOR values")
print(f"   ✅ Data placement: Based on PRED values")

# Create all possible columns from annotator values
all_annotator_columns = sorted(expanded_test_df['annotator'].unique())
unique_images = sorted(expanded_test_df['image_name'].unique())

# Initialize result dataframe
test_pivot = pd.DataFrame({'image_name': unique_images})
for col in all_annotator_columns:
    test_pivot[col] = 'NOT_FOUND'

print(f"   📊 Created columns: {list(test_pivot.columns)}")

# Group by image and pred for data placement
pred_to_words = (expanded_test_df.groupby(['image_name', 'pred'])['words']
                .apply(lambda x: ' | '.join(x) if len(x) > 1 else x.iloc[0])
                .reset_index())

print(f"   🔄 Processing pred-to-column placement...")

# Place words based on pred values
placements_made = 0
unmatched_preds = set()

for _, row in pred_to_words.iterrows():
    image_name = row['image_name']
    pred_value = row['pred']
    words_content = row['words']
    
    if pred_value in all_annotator_columns:
        # Place in matching column
        mask = test_pivot['image_name'] == image_name
        if mask.any():
            idx = test_pivot[mask].index[0]
            current_content = test_pivot.loc[idx, pred_value]
            if current_content == 'NOT_FOUND':
                test_pivot.loc[idx, pred_value] = words_content
                placements_made += 1
            else:
                test_pivot.loc[idx, pred_value] = f"{current_content} | {words_content}"
                placements_made += 1
    else:
        unmatched_preds.add(pred_value)

print(f"\n📊 CORRECTED Pivot Results:")
print(f"   Result shape: {test_pivot.shape}")
print(f"   Columns: {list(test_pivot.columns)}")
print(f"   Data placements made: {placements_made}")
if unmatched_preds:
    print(f"   Unmatched preds: {sorted(unmatched_preds)}")

print(f"\n📄 Test pivot result:")
print(test_pivot.to_string(index=False))

print(f"\n🔗 CORRECTED Logic Examples:")
print("=" * 70)
for _, example in pred_to_words.iterrows():
    pred_val = example['pred']
    words_preview = str(example['words'])[:30] + "..." if len(str(example['words'])) > 30 else str(example['words'])
    if pred_val in all_annotator_columns:
        print(f"   ✅ pred='{pred_val}' → placed in column='{pred_val}' | words='{words_preview}'")
    else:
        print(f"   ❌ pred='{pred_val}' → NO MATCHING COLUMN (not placed) | words='{words_preview}'")

print(f"\n✅ CORRECTED comma-separated handling works perfectly!")
print(f"   ✅ Columns created from: ANNOTATOR values (complete taxonomy)")
print(f"   ✅ Data distributed by: PRED values (model predictions)")
print(f"   ✅ Shows how model predictions map to ground truth structure")
print(f"   ✅ Reveals prediction bias and coverage gaps")
```
